## Concate Data Frames

## Merge DataFrames

```python
temp_df=pd.DataFrame({
    "city":["Hydderbad","Visakapatnam","Chennai","Bangalore"],
    "temperature":[32,45,30,34]
})
temp_df
```

```python
humidity_df=pd.DataFrame({
    "city":["Hydderbad","Kakinada","Chennai","Bangalore"],
    "humidity":[32,45,30,34]
})
humidity_df
```

```python
df=pd.merge(temp_df,humidity_df,on="city")
df
```

```python
df=pd.merge(temp_df,humidity_df,on="city",how="outer")
df
```

```python
df.isnull() #To find null values /missing values
```

```python
df.isnull().sum()
```

```python
df['temperature']=df['temperature'].fillna(df.temperature.min())
df
```
